**1) Ejercicio de herencia múltiple y polimorfismo:**
* Crea la clase Vehiculos con un constructor que incluya marca y modelo
* Esta clase Vehiculos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar gasolina"
* Crea la clase VElectricos con un constructor que incluya marca, modelo y autonomia
* Esta clase VElectricos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar electricidad"
* Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos, pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
* Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)
* Crear un objeto de BicicletaElectrica y otro de Quad
* Emplear las técnicas de polimorfismo aprendidas para conseguir que al peguntarle a los objetos que acabamos de crear por su repostaje, ambos nos respondan adecuadamente



```python
# Completa el ejercicio aquí
class Vehiculos:
    
    def __init__(self,marca,modelo):
        self.__marca = marca
        self.__modelo = modelo
    
    @ property
    def marca(self):
        return self.__marca
    @marca.setter
    def marca(self,new):
        self.__marca = new
    @ property
    def modelo(self):
        return self.__modelo
    @modelo.setter
    def modelo(self,new):
        self.__modelo= new
        
    def repostar(self):
        print("\n\t>>>repostando energia de tu Vehiculo<<<")
    
    def __str__(self):
        return """\n>>VEHICULOS<<
                MARCA: {}
                MODELO:{}""".format(self.marca,self.modelo)
class VElectricos:
    def __init__(self, marca, modelo,autonomia):
        self.__marca = marca
        self.__modelo = modelo
        self.__autonomia = autonomia
      
    @ property
    def marca(self):
        return self.__marca
    @marca.setter
    def marca(self,new):
        self.__marca = new
    @ property
    def modelo(self):
        return self.__modelo
    @modelo.setter
    def modelo(self,new):
        self.__modelo= new
    @ property
    def autonomia(self):
        return self.__autonomia
    @autonomia.setter
    def autonomia(self,new):
        self.__autonomia = new
        
        
    def repostar(self):
        print("\n\t>>>repostando energia de tu Vehiculo Electrico<<<")
    
    def __str__(self):
        return """\n>>VEHICULOS<<
                MARCA: {}
                MODELO: {}
                AUTOMONIA: {}""".format(self.marca,self.modelo,self.autonomia)
            
class vicicleta_electrica(VElectricos,Vehiculos):
    def __init__(self, marca, modelo, autonomia):
        VElectricos.__init__(self,marca, modelo, autonomia)
        Vehiculos.__init__(self,marca,modelo)
        print("has agregado tu Vehiculo",self.marca)
        
class Quad(Vehiculos,VElectricos):
    def __init__(self, marca, modelo,autonomia):
        super().__init__(marca, modelo)
        VElectricos.__init__(self,marca,modelo,autonomia)
        print("has agregado tu Vehiculo",self.marca)
        
#vehiculos

# vicicletas electricas        
vici1= vicicleta_electrica(marca="superVici",modelo="Infantil",autonomia= 50)
print(vici1)
vici1.repostar()

q1 = Quad(marca="Motorx",modelo="XS",autonomia=100)
print(q1)
q1.repostar()
print(q1.autonomia)

            
```

```python
# Completa el ejercicio aquí
class Personal_Universitario:
    
    def __init__(self, id, nombre, email):
        self.__id = id
        self.__nombre = nombre
        self.__email = email
        
    @ property
    def id (self):
        return self.__id 
    @id.setter
    def id (self,new):
        self.__id = new    
    @ property
    def nombre (self):
        return self.__nombre
    @nombre.setter
    def nombre (self,new):
        self.__nombre= new
    @ property
    def email (self):
        return self.__email
    @email.setter
    def email (self,new):
        self.__email= new    
        
    def __str__(self):
        return """\n>>PERSONAL<<
                id: {}
                NOMBRE: {}
                EMAIL: {}""".format(self.id,self.nombre,self.email)
                
class Oficina(Personal_Universitario):
    def __init__(self, id, nombre, email,puesto):
        super().__init__(id, nombre, email)
        self.__puesto = puesto
    @ property
    def puesto (self):
        return self.__puesto
    @puesto.setter
    def puesto (self,new):
        self.__puesto= new
        
    def __str__(self):
        return super().__str__() + "\n\t\t puesto: {}".format(self.puesto)
                     

class Profesor(Personal_Universitario):
    def __init__(self, id, nombre, email,especializacion):
        super().__init__(id, nombre, email)
        self.__especializacion = especializacion
        
    @ property
    def especializacion (self):
        return self.__especializacion
    @especializacion.setter
    def especializacion (self,new):
        self.__especializacion= new 
    def __str__(self):
        return super().__str__() + "\n\t\t especializacion:{}".format(self.especializacion)    

class Estudiante(Personal_Universitario):
    def __init__(self, id, nombre, email,CreditosAprovados):
        super().__init__(id, nombre, email)
        self.__CreditosAprovados = int(CreditosAprovados)
        
    @ property
    def CreditosAprovados(self):
        return self.__CreditosAprovados
    @CreditosAprovados.setter
    def CreditosAprovados (self,new):
        self.__CreditosAprovados= new 
    
    def __str__(self):
        return super().__str__() + "\n\t\t<<<creditos>>>: {}".format(self.CreditosAprovados)

def Mostrarclases(personas):
    for i in personas:
            print(i)

p1 = Estudiante("0145","Santiago Rodriguez","santirodriguz4@gmail.com",56)
p2 = Oficina("0280","Serguio Ramos","sergiorm23@gmail.com","Coordinador Estudiantil")
p3 = Profesor("0354","Maria Lucia Marquez","Luciamarquez50@hotmail.com","Ciencias Humanas")
p4 = Estudiante("0456","laura sofia gutierrez","laumr@gmail.com",85)
p5 = Oficina("0510","camilo duarte","camiloduarte@icloud.com","Secretario")
p6 = Profesor("0696","Maria Lucia Marquez","Luciamarquez50@hotmail.com","Lenguas Modernas")
personas = [p1,p2,p3,p4,p5,p6]
#mostrando clases
Mostrarclases(personas)
```
